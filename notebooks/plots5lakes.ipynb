```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os

# File to Load 
path1 = os.path.join("..", "data_files", "clean_data", "ice_coverage.csv")
path2 = os.path.join("..", "data_files", "api_calls", "Erie_22-27_GSOM.csv")
path3 = os.path.join("..", "data_files", "api_calls", "Huron_22-27_GSOM.csv")
path4 = os.path.join("..", "data_files", "api_calls", "Michigan_22-27_GSOM.csv")
path5 = os.path.join("..", "data_files", "api_calls", "Ontario_22-27_GSOM.csv")
path6 = os.path.join("..", "data_files", "api_calls", "Superior_22-27_GSOM.csv")

# Read the Data 
df1 = pd.read_csv(path1)
df2 = pd.read_csv(path2)
df3 = pd.read_csv(path3)
df4 = pd.read_csv(path4)
df5 = pd.read_csv(path5)
df6 = pd.read_csv(path6)
```

```python
df1["Period"] = df1["Year"].map(str) + "-" + df1["Month"].map(str)
df1.head()
```

```python
df2["period"] = df2['date'].str[:7]
df3["period"] = df3['date'].str[:7]
df4["period"] = df4['date'].str[:7]
df5["period"] = df5['date'].str[:7]
df6["period"] = df6['date'].str[:7]
```

### LAKE ERIE

```python
plt.figure(figsize=(10,7))
lake_erie = df1["Lake Erie"]
periods = df1["Period"]
plt.xlabel('Months')
plt.ylabel('Max % Ice Coverage')
plt.ylim(0,105)
plt.xticks(rotation = 90)
plt.scatter(periods, lake_erie)
plt.show()
```

```python
plt.figure(figsize=(10,7))
lake_erie2 = df2.groupby("period").agg({"value":["median"]})
periods2 = df2["period"]
plt.xlabel('Months')
plt.ylabel('No. of days with max temp <32 F')
plt.ylim(0,32)
plt.xticks(rotation = 90)
plt.scatter(lake_erie2.index, lake_erie2)
plt.show()
```
