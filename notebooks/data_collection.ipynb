```python
import pandas as pd
import os
import requests
import datetime as dt
import json
from config import noaa_token as token
```

```python
#variables
base = 'https://www.ncdc.noaa.gov/cdo-web/api/v2/'
header = {'token':token,
          'Content-Type':'application/json'}

state_ids = {'PA':'FIPS:42',
            'OH':'FIPS:39',
            'MI':'FIPS:26',
            'IL':'FIPS:17',
            'WI':'FIPS:55',
            'MN':'FIPS:27',
            'IA':'FIPS:18',
            'NY':'FIPS:36'}

variables = {'ice_thinkness':'THIC',
            'precipitation':'PRCP',
            }

payload = {'startdate':'2000-01-01',
          'enddate':'2000-12-31',
          'datasetid':'GHCND',
          'locationid':'FIPS:55',
          'includeAttributes':'true',
          'limit':'1000',
          'offset':'0'}
files_dir = os.path.join('..','data_files','api_data')
```

```python
#get count
def get_c(json):
    return json['metadata']['resultset']['count']

def is_not_file(filename):
        if os.path.isfile(filename):
            return False
        else:
            return True
```

```python
def gather(endpt,load):
    count = int(load['limit'])
    while int(load['offset']) < count:
        filename = load['locationid']+'_'+load['startdate'][:4]+'_'+load['offset']+'.csv'
        if is_not_file(os.path.join(files_dir,filename)):
            r = requests.get(base+endpt,headers=header,params=load).json()
            results = r['results']
            pd.DataFrame(results).to_csv(os.path.join(files_dir,filename))
            count = get_c(r)
            print(str(int(load['offset'])/int(count)*100)+'% Complete')
        else:
            print(f"Already Done! {str(int(load['offset']))} % Complete")
            count = count + int(load['limit'])
        load['offset'] = str(int(load['offset'])+int(load['limit']))
    print(f"100% Complete! There are now {load['offset']} files in {files_dir} for {load['locationid']}, {load['startdate']} through {load['enddate']}")
    return
```

```python
gather('data',payload)
```
